```python
import pandas as pd
import matplotlib
from single_var_euler_estimator import EulerEstimator
```

```python
def derivative(x):
    return x - 2
```

```python
euler = EulerEstimator(derivative)
test1 = pd.DataFrame(euler.calc_estimated_points((0, -2), 0.25, 10))
test1_plot = test1.plot(kind="scatter", x=0, y=1,
                        color="red", figsize=(10, 10), label="y(0) = -2")

test2 = pd.DataFrame(euler.calc_estimated_points((0, -1), 0.25, 10))
test2_plot = test2.plot(kind="scatter", x=0, y=1, ax=test1_plot, color="green", label="y(0) = -1")

test3 = pd.DataFrame(euler.calc_estimated_points((0, 0), 0.25, 10))
test3_plot = test3.plot(kind="scatter", x=0, y=1, ax=test2_plot, color="blue", label="y(0) = 0")

test4 = pd.DataFrame(euler.calc_estimated_points((0, 1), 0.25, 10))
test4_plot = test4.plot(kind="scatter", x=0, y=1, ax=test3_plot, color="yellow", label="y(0) = 1")

test5 = pd.DataFrame(euler.calc_estimated_points((0, 2), 0.25, 10))
test5.plot(kind="scatter", x=0, y=1, ax=test4_plot, color="purple", label="y(0) = 2")
```
